# pie chart

# univariate and  bivarte analysis
#  numerical ->categorial
# use case : to find  contribution on standard scale

```python
dates = [23,45,100,20,45]
subjects= ['eng','urdu','maths','islamyait','pashto']
plt.pie(dates,labels= subjects,autopct='%0.1f%%',colors=['blue','green','yellow','gray','red'])
plt.show()
```

```python
plt.pie(dates,labels=subjects,autopct='%0.1f%%',explode=[0.1,0,0,0,0.1])
plt.show()
```

```python
fig, ax =  plt.subplots(nrows=2,ncols=1,sharex=True)
ax[0].scatter(df['gpa'],df['plament'])
ax[1].scatter(df['iq'],df['plament'])
ax[0].set_title('student table')
ax[0].set_ylabel('gpa')
ax[1].set_title('student table')
ax[1].set_ylabel('gpa')




```

# 3d satter plots

```python
fig = plt.Figure()
ax = plt.subplot(projection='3d')


```

```python
a= [0,1,5,25]
b=[0,10,13,0]
d=[0,13,20,9]
fig = plt.figure()
ax = plt.subplot(projection='3d')
ax.scatter3D(a,b,d ,s=[100,100,100,100])
ax.plot3D(a,b,d,color='red')

```

```python
a = np.linspace(-10,10.100)
b=np.linspace(-10,10,100)
aa,bb = np.meshgrid(a,b)
t = aa**2 + bb**2

t.shape
```

```python
fig =plt.Figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(aa,bb,t,cmap='viridis')
fig.colorbar(p)
```
